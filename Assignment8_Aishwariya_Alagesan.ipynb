#### Qn 1

```python
import numpy as np
import matplotlib.pyplot as plt

random_numbers = np.random.randint(1, 101, size=10000)
```

```python
sample_means = []

for i in range(1000):
    # Randomly sample 30 numbers from the population
    random_sample = np.random.choice(random_numbers, size=30)

    random_sample_mean = np.mean(random_sample)
    sample_means.append(random_sample_mean)
```

```python
plt.hist(sample_means, bins=30, edgecolor='k', alpha=0.7)
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.title('Distribution of 1,000 Sample Means')
plt.show()
```

The distribution of sample means will tend to be approximately normally distributed, even if the population distribution is not normally distributed. 

#### Qn 2

```python
import numpy as np

population_exp = np.random.exponential(2, size=10000)
```

```python
sample_means_exp = []

for i in range(500):
    # Randomly sample 50 numbers from the population
    random_sample_exp = np.random.choice(population_exp, size=50)

    random_sample_mean_exp = np.mean(random_sample_exp)
    sample_means_exp.append(random_sample_mean_exp)
```

```python
plt.hist(sample_means_exp, bins=30, edgecolor='k', alpha=0.7)
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.title('Distribution of 500 Sample Means')
plt.show()
```

The Central Limit Theorem (CLT) applies to the sampling of 10,000 numbers from an exponential distribution, ensuring that the distribution of sample means becomes approximately normal with a mean close to the population mean and reduced variability as the sample size increases, providing robust statistical inference capabilities.

#### Qn 3

```python
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Given values
sample_std = np.std(sample_means_exp, ddof=1)
sample_size = 50
confidence_level = 0.95

# Z-score for 95% confidence
z_score = norm.ppf((1 + confidence_level) / 2)

# Calculate margin of error
margin_of_error = z_score * (sample_std / math.sqrt(sample_size))

# Calculate confidence interval
lower_limit = np.mean(sample_means_exp) - margin_of_error
upper_limit = np.mean(sample_means_exp) + margin_of_error

# Output the confidence interval
print(f"The 95% confidence interval for the population mean is ({lower_limit:.2f}, {upper_limit:.2f})")
```

A 95% confidence interval of (1.91, 2.05) for the population mean suggests that we can be 95% confident that the true population mean falls within this interval. In other words, it provides a range of values within which we believe the population mean is likely to be.

#### Qn 4

```python
import numpy as np
import matplotlib.pyplot as plt

population = np.random.randint(1, 101, 1000)

# Lists to store sample means
sample_means_10 = []
sample_means_30 = []

# Perform 1,000 simulations for each sample size
for _ in range(1000):
    # Randomly sample 10 numbers and calculate the sample mean
    sample_10 = np.random.choice(population, 10)
    sample_mean_10 = np.mean(sample_10)
    sample_means_10.append(sample_mean_10)

    # Randomly sample 30 numbers and calculate the sample mean
    sample_30 = np.random.choice(population, 30)
    sample_mean_30 = np.mean(sample_30)
    sample_means_30.append(sample_mean_30)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

ax[0].hist(sample_means_10, bins=30, edgecolor='k', alpha=0.7)
ax[0].set_title('Distribution of Sample Means (Sample Size 10)')
ax[0].set_xlabel("Sample Mean")
ax[0].set_ylabel("Frequency")

ax[1].hist(sample_means_30, bins=30, edgecolor='k', alpha=0.7)
ax[1].set_title('Distribution of Sample Means (Sample Size 30)')
ax[1].set_xlabel("Sample Mean")
ax[1].set_ylabel("Frequency")

plt.show()
```

This observation aligns with the Central Limit Theorem (CLT), which states that as the sample size increases, the sampling distribution of the sample mean becomes closer to a normal distribution with a smaller standard deviation.
