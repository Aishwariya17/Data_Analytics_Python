#### Qn 4

```python
data['time'] = pd.to_datetime(data['time'])

data['year'] = data['time'].dt.year
data['month'] = data['time'].dt.month
data['day'] = data['time'].dt.day

data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

year_counts = data['year'].value_counts().sort_index()

month_counts = data['month'].value_counts().sort_index()

fig, (plt1, plt2) = plt.subplots(1, 2, figsize=(8,4))
sns.barplot(x=year_counts.index, y=year_counts.values, ax=plt1)
plt1.set_title("Earthquakes over Year")
plt1.set_xlabel("Year")
plt1.set_ylabel("Count")

sns.barplot(x=month_counts.index, y=month_counts.values, ax=plt2)
plt2.set_title("Earthquakes over Month")
plt2.set_xlabel("Month")
plt2.set_ylabel("Count")

plt.tight_layout()
plt.show()
```

#### Qn 5

```python
plt.figure(figsize=(8, 4))
sns.histplot(data['mag'])
plt.title("Distribution of Earthquake Magnitudes")
plt.xlabel("Magnitude")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=data, y='mag')
plt.title("Visualization of distribution of Earthquake Magnitudes")
plt.ylabel("Magnitudes")
plt.show()
```

#### Qn 6

```python
plt.figure(figsize=(7, 4))
sns.histplot(data['depth'], bins=40)
plt.title("Distribution of Earthquake Depths")
plt.xlabel("Depth")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(7, 4))
sns.boxplot(data=data, x='depth')
plt.title("Visualization of Earthquake Depths")
plt.xlabel("Depth")
plt.show()
```

#### Qn 7

```python
location_source = data['locationSource'].value_counts()

top_10_location = location_source.head(10).reset_index()
top_10_location.columns = ['Location', 'No of Earthquakes']

top_10_location
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='No of Earthquakes', y='Location', data=top_10_location, palette='Greens')
plt.xlabel("No of Earthquakes")
plt.ylabel("Location")
plt.title("Top 10 location with highest no of earthquakes")
plt.show()
```

#### Qn 8

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(x='mag', y='depth', data=data, alpha=0.8)
plt.xlabel("Mag")
plt.ylabel("Depth")
plt.title("Magnitude vs Depth")

corr = data['mag'].corr(data['depth'])
plt.text(7.5, 1.1, f'Correlation Coefficient: {corr:.2f}', fontsize=12, color='green')

plt.show()
```

#### Qn 9 

```python
plt.figure(figsize=(10, 5))
plt.scatter(data['longitude'], data['latitude'], c=data['mag'], cmap='viridis', s=data['mag'] * 10, alpha=0.7)
plt.colorbar(label='Magnitude')
plt.title("Earthquake Locations and Magnitudes")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)

plt.show()
```
