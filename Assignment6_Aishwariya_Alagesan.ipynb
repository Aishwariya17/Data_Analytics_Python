```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

file_path = 'C:/Users/Aishwariya/Downloads/fb_stock_prices_2018.csv'
data1 = pd.read_csv(file_path)
df = pd.DataFrame(data1)

dates = pd.to_datetime(df['date'])
close_price = df['close']

start_date = '2018-07-25'
end_date = '2018-07-31'

plt.figure(figsize=(7, 4))
plt.plot(dates, close_price, label='dose', color='blue')

plt.axvspan(start_date, end_date, color='blue', alpha=0.3, label='July decline')

plt.xlabel('date')
plt.ylabel('$Price')
plt.title('Facebook Closing Price')
plt.legend()

plt.show()
```
